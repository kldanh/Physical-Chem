`plt.plot(x,y,'o')`

## Creating Arrays for Plotting
We can create an array of evenly spaced numbers for our x-axis two ways:  `arange`, which you saw in a previous worksheet, and `linspace`.
### Arange
The command `np.arange(start, end, step)` creates an array that starts at the start value, increases by the value of step, and ends *before* it gets to the end value.  For example, `np.arange(5,15, 2)` would create the array `[5, 7, 9, 11, 13]
### Linspace  
NumPy's linspace function allows us to create a range of evenly spaced values over a specified interval using this syntax: `np.linspace(start, end, number of values)`. Note that this time, we specify N, the number of points, *not* the step value, and that the end value *is* included. For example, `np.linspace(15,25, 5)` would create the array `[15, 17.5, 20, 22.5, 25]'.



```python


x1 = np.linspace(0,5,11) # Create 11 points evenly spaced from 0 to 5
print(x1) # Display the x1 array
y1 = 2*x1 # You can create a new array doing math on them. This will multiply each value in the x1 array by 2 and store the result in y1.
print(y1) # Display the y1 array. Note that it's double the x1 array. Try changing the formula for y1.

# Let's use it to plot something interesting
x2 = np.linspace(0, 2 * np.pi, 100)
y2 = np.sin(x2)
plt.plot(x2,y2,'-') # Here we have dense, quasi-continuous data, so it makes sense to plot it as a smooth curve rather than with markers.
```
